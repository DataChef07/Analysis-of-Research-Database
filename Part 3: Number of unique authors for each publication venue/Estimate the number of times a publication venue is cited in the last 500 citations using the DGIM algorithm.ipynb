```python
import pandas as pd
```

```python
# load the data
l1 = []
with open(r'C:/Users/hp/Big Data Analytics (BDA)/Assignment3_question2_dataset.txt', 'r') as fp:
    for line in fp:
        x = line[:-1]
        l1.append(x)

print(l1)
```

```python
#take last 500 elements
l2 = l1[-500:]
```

```python
#find the unique items
s = set(l2)
len(s)
```

```python
#check count of each unique elements
count = pd.Series(l2).value_counts()
count
```

```python
#run DGIM
#  For this, we need a bit vector(with True/False values). Hence, converting the original list ‘l1’ into a bit vector.
from dgim import Dgim
dgim = Dgim(N=500, error_rate=0.1) #We have to check for the last 500 elements hence, set N=500.
for i in l1:
    #For the item in which we want to find the occurrences, if it is present in ‘l1’ then update the bit vector with ‘True’. 
    if i=="Combinatorica":
        dgim.update(True)
    else:
        dgim.update(False) # For anything other than the searched item, update the vector as ‘False’.
dgim_result = dgim.get_count()   #get the count
```
